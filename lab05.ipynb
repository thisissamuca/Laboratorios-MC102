```python
##################################################
# MC102 - Algoritmos e Programação de Computadores
# Laboratório 5 - Jornada de Trabalho
# Nome: Samuel Augusto de Souza
# RA: 261025
##################################################

periodos = []
periodos.clear()

horas_totais = []
horas_totais.clear()

horario = []
horario.clear()

horas_extras = []
horas_extras.clear()

def salario():

# Leitura do valor da hora

  V = int(input()) # Valor hora

# Leitura do numero de dias trabalhados na semana

  D = int(input()) # Número de dias trabalhados

# Leitora e processamento dos periodos de trabalho de cada dia

  for i in range (D):

    qtdPeriodo = int(input())

    if qtdPeriodo > 1:

      for j in range(qtdPeriodo):

        hora_inicio = int(input())

        hora_fim = int(input())

        horario.append(hora_fim - hora_inicio)

      if sum(horario) > 8:

        horas_extras.append(sum(horario) - 8)

      if sum(horario) <= 8:

        valor_horas_extras = 0

      valor_horas_extras = (V/2 * sum(horas_extras))

      periodos.append(sum(horario))

      horario.clear()

    else:

      hora_inicio = int(input())

      hora_fim = int(input())

      periodos.append(hora_fim - hora_inicio)

      if periodos[len(periodos) - 1] > 8:

        horas_extras.append(periodos[len(periodos) - 1] - 8)

      if periodos[len(periodos) - 1] <= 8:

        valor_horas_extras = 0

      valor_horas_extras = (V/2 * sum(horas_extras))

  horas_totais.append(sum(periodos))

  if horas_totais[len(horas_totais) - 1] > 44:

    if horas_extras == []:

      horas_extras.append(horas_totais[len(horas_totais) - 1] - 44)

    if horas_totais[len(horas_totais) - 1] - 44 > sum(horas_extras):

      horas_extras.clear()

      horas_extras.append(horas_totais[len(horas_totais) - 1] - 44)

    valor_horas_extras = (V/2 * sum(horas_extras))

  valor_horas_trabalhadas = (horas_totais[len(horas_totais) - 1] * V)

# Calculo do valor devido ao funcionário

  valor = valor_horas_trabalhadas + valor_horas_extras

# Impressão da saída

  print("Horas trabalhadas:", horas_totais[len(horas_totais) - 1])

  print("Horas extras:", sum(horas_extras))

  print("Valor devido: R$ {:0.2f}".format(valor))

salario()
```
