```python
###################################################
# MC102 - Algoritmos e Programação de Computadores
# Laboratório 7 - Disconnect
# Nome: Samuel Augusto de Souza
# RA: 261025
###################################################

# Leitura das listas

lista = []
lista02 = []
lista03 = []
lista04 = []
lista05 = []
lista06 = []

# Leitura das tropas de defesa

n = int(input())

tupla = ()

for i in range(n):

  x = int(input())

  tupla = tupla + tuple([x])

# Leitura das tropas de ataque

m = int(input())

tupla02 = ()

for j in range(m):

  y = int(input())

  tupla02 = tupla02 + tuple([y])

# Processamento da guerra

for i in range(len(tupla)):

  if len(tupla[0 + i:len(tupla02) + i]) >= len(tupla02):

    lista.append(tupla[0 + i:len(tupla02) + i])

for j in range(len(lista)):

  for k in range(len(lista[j])):

    lista02.append(lista[j][k] - tupla02[k])

for l in range(1, len(lista)):

  lista03.append(lista02[0 + (l-1)*m:m + (l-1)*m])

for m in range(len(lista03)):

  lista04.append([])

  for n in range(len(lista03[m])):

    if lista03[m][n] < 0:
      
      lista04[m].append(-1)

    if lista03[m][n] >= 1:
      
      lista04[m].append(1)

    if lista03[m][n] == 0:
      
      lista04[m].append(0)

for o in range(len(lista04)):

  lista05.append(sum(lista04[o]))

# Obtendo o posicionamento das tropas

for p in range(len(lista05)):

  if lista05[p] < 0:

    lista06.append(lista05.index(lista05[p]) + 1)

# Saída de dados

if lista06 == []:

  print('Derrota')

else:

  posicao = lista06[0]
  
  print('Vitória posicionando as tropas a partir da posição', posicao)
```
